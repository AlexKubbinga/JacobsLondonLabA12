```python
# Importing the required libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup


# In[21]:


# Downloading contents of the web page
url = "https://www.dreso.com/de/en/career/job-opportunities"
data = requests.get(url).text


# In[22]:


# Creating BeautifulSoup object
soup = BeautifulSoup(data, 'html.parser')
#print(soup.prettify())


```

```python
# In[25]:


job_id_line = soup.find_all('td', 'title job-item')
print(len(job_id_line))
job_id_line[0]


```

```python
# In[31]:

job_ids=[]
import re
for i in range(len(job_id_line)):
    job_id_each=re.search('Vacancies/(\d+)/Description', str(job_id_line[i]))[1]
    job_ids.append(job_id_each)
```

```python
df['Job ID']=job_ids
df.rename(columns = {'Entry level':'Job_tittle', 'Time limit':'Full_or_part_time','Entry as':'level_of_seniority','Job ID':'Job_ID'}, inplace = True)
```
